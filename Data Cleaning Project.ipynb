<h1>Objective 3: To correct invalid data entries</h1>

<h4>For "Name" Column</h4>

```python
df["Name"]=df["Name"].str.strip()
```

```python
#Removing numeric names
df["Name"]=df["Name"].replace(r'^\d+$',np.nan,regex=True)
```

```python
df["Name"]
```

```python
pd.isnull(df['Name']).sum()
```

<h4>For "Age" Column</h4>

```python
#Converting "twenty" to 20
df["Age"]=df["Age"].replace("twenty",20)
```

```python
df[df["Age"]==20]
```

```python
df["Age"] = pd.to_numeric(df["Age"],errors="coerce")
```

```python
#Age less than 0 and greater than 100 becomes nan
df.loc[(df["Age"]<0) | (df["Age"]>100), "Age"]=np.nan
```

```python
df["Age"]
```

<h4>For "Gender" Column</h4>

```python
df["Gender"]=df["Gender"].replace({"M":"Male","F":"Female","Unknown":np.nan})
```

```python
df["Gender"]
```

<h4>For "Marks" Column</h4>

```python
df["Marks"]=df["Marks"].replace({"abc":np.nan,"Ninety":90,"absent":np.nan,"N/A":np.nan})
```

```python
df["Marks"]=pd.to_numeric(df["Marks"],errors="coerce")
```

```python
df["Marks"]
```

<h4>For "Branch" Column</h4>

```python
df["Branch"]=df["Branch"].str.upper()
```

```python
df["Branch"]
```

```python
#Removing dots in branches
df["Branch"]=df["Branch"].str.replace(".","",regex=False)
```

```python
df["Branch"]
```

```python
#Removing white spaces in Branches
df["Branch"]=df["Branch"].str.replace(" ","",regex=False)
```

```python
df["Branch"]
```

<h4>For "Email" Column</h4>

```python
#General form of an email
p="^[\w]+@[\w]+\.[\w]+$"
```

```python
df['Email']=df['Email'].where(df['Email'].str.match(p), 'invalid')
```

```python
df["Email"]
```

```python
df['Email']=df['Email'].replace(" ", np.nan)
```

<h4>For "Join_Date" Column</h4>

```python
df["Join_Date"]=pd.to_datetime(df["Join_Date"], errors="coerce", dayfirst=True)
```

<h4>For "Fees_Paid" Column</h4>

```python
df["Fees_Paid"]=df["Fees_Paid"].replace({"forty thousand":40000})
```

```python
df["Fees_Paid"]=pd.to_numeric(df["Fees_Paid"],errors="coerce")
```

```python
df.loc[(df["Fees_Paid"]<0),"Fees_Paid"]=np.nan
```
